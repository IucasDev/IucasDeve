CONHECENDO STR

Algumas funções úteis com .str:
Função O que faz

.str.upper() Tudo maiúsculo

.str.lower() Tudo minúsculo

.str.title() Primeira letra de cada palavra em maiúsculo

.str.contains("texto") Verifica se contém certo texto

.str.replace("a", "b") Substitui "a" por "b"

.str.strip( Remove espaços em branco
