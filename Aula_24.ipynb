```python
# Agrupa os clientes por estado e soma os gastos
gastos_por_estado = df.groupby('Estado')['Gasto_Total'].sum()

# Filtra apenas os estados com gasto total acima de 5000
gastos_acima_5000 = gastos_por_estado[gastos_por_estado > 5000]

# Mostra o resultado
gastos_acima_5000
```
